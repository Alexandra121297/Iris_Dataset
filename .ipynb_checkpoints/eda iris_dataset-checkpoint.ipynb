```python
#Se importa librerías de análisis de datos y visualización

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns  
```

```python
#Variable para la rota, se lee correctamente, se genera el índice de manera automática por pandas

path_to_data = "./iris_dataset.csv"
df = pd.read_csv(path_to_data)
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
#Se especifica los tipos de variables

categoricas = ["species"]
numericas = ["sepal_length", "sepal_width", "petal_length", "petal_width"]
```

```python
df[numericas].describe()
```

```python
#Información por columna

for col in numericas:
    print(f"Característica: {col}")
    print(df[col].value_counts())
    print()
```

```python
#Se relaciona la longitud y el ancho del sépalo y se discrimina por especies

sns.scatterplot(data=df, x="sepal_width", y="sepal_length", hue="species")
```

```python
#Se relaciona la longitud y el ancho del pétalo y se discrimina por especies

sns.scatterplot(data=df, x="petal_width", y="petal_length", hue="species")
```
