# 3. Gráficos
Objetivo: Determinar las especies por las siguientes características: sepal_length, sepal_width, petal_length, petal_width. 

+ ### Barptlos

```python
sns.catplot(x ="species", hue ="species", kind ="count", data = df)
```

Se observa:
+ Que, las tres especies poseen la misma cantidad de plantas.


+ ### Scatterplot

```python
#Se relaciona la longitud y el ancho del sépalo, se discrimina por especies.

sns.scatterplot(data=df, x="sepal_width", y="sepal_length", hue="species")
```

Se observa:
+ La especie Iris-setosa posee sépalos que tienden a ser más anchos, entre 2.9 a 4.5 cm; que largo, máxima 5.7 cm aprox.. Se identifica un outlier dentro de las setosas. 
+ La especie Iris-versicolor posee sépalos con mayor longitud, máximo 7.0 cm; y, menor ancho, máximo 3.4 cm aprox. Se identifica un outlier dentro de las versicolor. 
+ La especie Iris-virginica posee sépalos que tiende a ser de mayor longitud, máximo 8.0 cm; y, de ancho entre los 2.2 y 3.9 cm aprox. Dudas sobre outliers. 

```python
#Se relaciona la longitud y el ancho del pétalo, se discrimina por especies.

sns.scatterplot(data=df, x="petal_width", y="petal_length", hue="species")
```

```python
sns.pairplot(df, hue="species", palette="rocket")
```

```python
sns.violinplot(x="species", y="petal_length", data=df)
```

```python
sns.boxplot(x="species", y="petal_width", data=df, hue="species")
```

```python
df_g=sns.PairGrid (data= df, hue="species",
                   x_vars= ["petal_length", "petal_width"],
                    y_vars= ["sepal_length", "sepal_width"])
df_g.map(plt.scatter)
df_g.add_legend()
```

```python
mask = np.triu(np.ones_like(df.corr(), dtype=bool))
f, ax = plt.subplots(figsize=(11, 9))
sns.heatmap(df.corr(), mask=mask, vmax=1., vmin=-1., center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5}, annot=True)
```
